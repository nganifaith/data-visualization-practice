## Part 1A
1) import the library needed for data importation

2) import the data set

```python
import numpy as np
import pandas as pd

df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
df.head()
```

## Part 1B

Carryout a quick data cleaning on the dataset

```python
df.columns
```

```python
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)

df.head()
```

```python
column_rename = {'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'}
```

```python
df.rename(columns = column_rename, inplace = True)

df.columns
```

### Part 1C

during abit of data manipulation to add a column Total, which indicates the total number of immigrates in each year

```python
df['Total'] = df.sum(axis = 1)
df.head()
```

```python
df.isnull().sum()
```

Setting the country column as the index of our dataset

```python
df = df.set_index('Country')
```

```python
df.head()
```

## Pie plot

to demostrate pie plot we will group the data into continents and plot their immigration ratios

```python
cont = df.groupby('Continent', axis=0).sum()
cont.head()
```

```python
cont['Total'].plot(kind='pie', figsize=(7,7),
                  autopct='%1.1f%%',
                  shadow=True)
#plt.title('Immigration By Continenets')
plt.axis('equal')
plt.show()
```

```python
## Custom plot colors and start angles

colors = ['lightgreen', 'lightblue', 'pink', 'purple', 'grey', 'gold']
explode=[0.1, 0, 0, 0, 0.1, 0.1]
cont['Total'].plot(kind='pie', figsize=(14, 8),
                  autopct = '%1.1f%%', startangle=90,
                  shadow=True, labels=None, pctdistance=1.00, colors=colors, explode = explode)
plt.axis('equal')
plt.legend(labels=cont.index, loc='upper right', fontsize=10)
plt.show()
```
